```python
def my_exp(x, e):
    print(f"e={e} and x={x} in my_exp")
    return x ** e
```

```python
def flip_exp(e, x):
    print(f"e={e} and x={x} in flip_exp")
    return my_exp(x, e)
```

```python
dat["a"].apply(flip_exp, x=3)
```

```python
def avg_3_apply(col):
    return (np.mean(col))
```

```python
dat.apply(avg_3_apply)
```

```python
def avg_3_apply(col):
    x = col[0]
    y = col[1]
    z = col[2]
    return (x + y + z) / 3
```

```python
dat.apply(avg_3_apply)
```

```python
@np.vectorize
def v_avg_2_mod_np(x, y):
    if (x == 20):
        return np.NaN
    else:
        return (x + y) / 2
```

```python
%%timeit
v_avg_2_mod_np(dat['a'], dat['b'])
```

```python
rng  = np.random.RandomState(3)

ts = pd.Series(
    rng.normal(0,1,1000),
    index= pd.date_range('1-1-2000',periods=1000)
)

ts
```

```python
rdf = pd.DataFrame(
    rng.normal(0,1,size=(1000,4)),
    index = ts.index,
    columns=['A','B','C','D']
)

rdf.head()
```

```python
rdf.cumsum()
```

```python
plt.rcParams['figure.figsize']=(19,6)
rdf.cumsum().plot();
plt.show()
```
